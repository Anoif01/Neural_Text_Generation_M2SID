## Dans ce notebook, nous allons tracer les courbes depuis des fichiers sauvegardés.

```python
valid_info_paths = '/content/drive/MyDrive/GM5/DEEP/Paper/experiment'
# valid_info_paths = '/content/drive/MyDrive/GM5/DEEP/Paper/NeuralTextGeneration-master/experiment_pred'
# files = [int(f) for f in os.listdir(valid_info_paths)]
# f = max(files)
# f=2
valid_loss_tot = []
valid_time = []
train_loss_tot = []
train_time = []

for f in [3]:
    valid_info_path = valid_info_paths + '/' + str(f)

    for fil in os.listdir(valid_info_path):
        try:
            if fil == 'valid_loss.txt':
                path = valid_info_path + '/' + fil
                valid_loss = []
                comput_time = []
                for line in open(path):
                    epo, loss, t = line.split('\t')
                    loss = loss.split(':')[1]
                    t = t.split(':')[1]
                    valid_loss.append(float(loss))
                    comput_time.append(float(t))
                valid_loss_tot.append(valid_loss)
                mt = np.mean(comput_time)
                valid_time.append(mt)
                

            if fil == 'train_loss.txt':
                path = valid_info_path + '/' + fil
                train_loss = []
                comput_time = []
                for line in open(path):
                    epo, loss, t = line.split('\t')
                    loss = loss.split(':')[1]
                    t = t.split(':')[1]
                    train_loss.append(float(loss))
                    comput_time.append(float(t))
                train_loss_tot.append(train_loss)
                train_time.append(np.mean(comput_time))


            # if fil == 'sample.txt':
            #     path = valid_info_path + '/' + fil
            #     for line in open(path):
            #         print(line)
        except:
            pass
```

```python
## show perplexity.
perplexity = []
for loss in valid_loss_tot:
  perplexity.append(np.exp(loss))
perplexity = np.array(perplexity).reshape(10,1)
plt.plot(range(1,11), perplexity, 'go-')
plt.xlabel('epoch')
plt.title("Perplexity")
plt.show()
```

## Calculons le score BLEU

```python
import numpy as np
from nltk.translate.bleu_score import sentence_bleu
import os
def split(sentence): 
    return (sentence.split())

def compute_BLEU(sentence, ref):
  reference = split(ref)
  # if sentence == '. \n':
  #     return 0
  # else:
  # if '-lrb-' in sentence:
  #     reference.append('-lrb-')
  # if '-rrb-' in sentence:
  #     reference.append('-rrb-')

  candidate = split(sentence)
  # print(reference)
  # print(candidate)
  BLEU_score = sentence_bleu([reference], candidate)
  return BLEU_score

# Bleu_tot = []
for f in [3]:
# f = 6
  BLEU_per_epoch = np.zeros(10)
  for epoch in range(1, 11):
  
    expt_result_path = '/content/drive/MyDrive/GM5/DEEP/Paper/experiment/%d'%f
#     expt_result_path = '/content/drive/MyDrive/GM5/DEEP/Paper/NeuralTextGeneration-master/experiment_pred/%d'%f
    reference = 'leonard shenoff randle -lrb- born february 12 , 1949 -rrb- is a former major league baseball player .'
    res_path = os.path.join(expt_result_path, 'sample%d.txt'%epoch)
#     res_path = os.path.join(expt_result_path, 'sample.txt')
    file =  open(res_path, "r")
    Bleu = 0.
    nb = 0
    for sentence in file:
      if sentence != '. \n':
        bleu = compute_BLEU(sentence, reference)
        # print("bleu: %4f" % bleu)
        Bleu += bleu
        nb += 1
    print('Bleu : ',  Bleu /  nb)
    BLEU_per_epoch[epoch-1] = Bleu /  nb
  Bleu_tot.append(BLEU_per_epoch)
  # print("BLEU metric for epoch %d : %4f" % (epoch, BLEU_per_epoch[epoch-1]))
```

## BLEU de n_gram

```python
plt.plot(Bleu_tot[0], 'bo-', label='10gram')
plt.plot(Bleu_tot[1], 'go-', label='11gram')
plt.plot(Bleu_tot[2], 'ro-', label='12gram')
plt.plot(Bleu_tot[3], 'ro-', label='14gram')
plt.title("Bleu score")
plt.legend()
plt.xlabel('epoch')
plt.show()

Bleus = np.mean(Bleu_tot, 1)
plt.plot(Bleus, 'bo-')
plt.title("Mean of Bleu Score of 10 epochs")
plt.xticks([0, 1, 2, 3], ['10','11','12', '14'], fontproperties='STKAITI')
plt.show()
```
